```python
import random
f=open("data.csv","w")
f.write("id,name,age,sal")
names=["samba","aneel","nithin","abhijith","munna","anand"]
age=range(23,30)
sal=[30000,35000,45000,65000,85000,95000]
for i in range(1000):
    row="\n%s,%s,%s,%s"%(i,random.choice(names),random.choice(age),random.choice(sal))
    f.write(row)
f.close()
```

```python
f=open("data.csv")
data=f.read()
salaries=[]
for row in data.splitlines():
    s=row.split(",")[-1]
    salaries.append(s)
salaries.remove("sal")
salaries_int=[]
for  sal in salaries:
    salaries_int.append(int(sal))
print(salaries_int)
```
