```python
persons=[{"name":"jay","id":1,"sal":12345},
        {"name":"sai","id":2,"sal":12346},
         {"name":"vedha","id":3,"sal":12347},
         {"name":"samba","id":4,"sal":12348},
        ]
f=open("data1.csv","w")
f.write(str(persons))
f.close()
```

```python
import pickle
persons=[{"name":"jay","id":1,"sal":12345},
        {"name":"sai","id":2,"sal":12346},
         {"name":"vedha","id":3,"sal":12347},
         {"name":"samba","id":4,"sal":12348},
        ]
f=open("data2.csv","wb")
pickle.dump(persons,f)
f.close()
```

```python
f=open("data2.csv","rb")
data=f.read()
print(data)
print(type(data))
```

```python
f=open("data2.csv","rb")
data=pickle.load(f)
print(data)
print(type(data))
```

```python
for i in data:
    print(i)
    break
```

```python
persons={"data":[{"name":"jay","id":1,"sal":12345},
        {"name":"sai","id":2,"sal":12346},
         {"name":"vedha","id":3,"sal":12347},
         {"name":"samba","id":4,"sal":12348}],
         "metadata":{"size":"25MB"}
        }
import yaml
f=open("data.yaml","w")
yaml.dump(persons,f)
f.close()
```

```python
f=open("data.yaml","r")
data=yaml.load(f)
print(data)
print(type(data))
```
